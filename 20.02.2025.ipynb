```python
plt.figure(figsize=(9,9))

plt.subplot(2, 2, 1)  # 1. grafik (1. satÄ±r, 1. sÃ¼tun)
plt.title("Grafik 1")

plt.subplot(2, 2, 2)  # 2. grafik (1. satÄ±r, 2. sÃ¼tun)
plt.title("Grafik 2")

plt.subplot(2, 2, 3)  # 3. grafik (2. satÄ±r, 1. sÃ¼tun)
plt.title("Grafik 3")

plt.subplot(2, 2, 4)  # 4. grafik (2. satÄ±r, 2. sÃ¼tun)
plt.title("Grafik 4")

plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.title("Grafik 1")
plt.plot(x, y, 'r')
plt.xlabel("X Label")
plt.ylabel('Y Label')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
plt.title("Grafik 2")
plt.ylabel('X Label')
plt.xlabel('Y Label')
plt.tight_layout()
```

```python
# OOM Object Oriented Method

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('X = Y^2 Graph')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.4, 0.3, 0.3])

axes1.plot(x, y)
axes1.set_xlabel('X Axis')
axes1.set_title('Big Plot')
axes2.plot(y, x)
axes2.set_ylabel('X Axis')
axes2.set_title('Small Plot')
```

```python
fig, axes = plt.subplots(1, 3)  # 1 satÄ±r, 3 sÃ¼tun subplot

axes[0].set_title("Birinci subplot")
axes[1].set_title("Ä°kinci subplot")
axes[2].set_title("ÃœÃ§Ã¼ncÃ¼ subplot")
plt.tight_layout()
plt.show()
```
